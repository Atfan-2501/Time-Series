Decomposing time series using statsmodels

```python
import pandas as pd
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("C:/Users/2068671/Downloads/UDEMY_TSA_FINAL/UDEMY_TSA_FINAL/Data/airline_passengers.csv", index_col = 0, parse_dates = True)
```

```python
df.columns
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
plt.figure()
df.plot(figsize = (10,5))
plt.ylabel("No of Passengers")
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df["Thousands of Passengers"], model= "multiplicative")
```

```python
result.resid.head(10)
```

Plotting the decomposed time series - i.e Residual, trend and seasonality
