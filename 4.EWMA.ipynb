Exponential Weighted Moving Average
1. The weights are in exponentially decreasing order where the most recent value gets the most weight
2. This helps overcoming some shortcomings of SMA like it reduces the lag effect

Span, centre of mass and Half-life is used to calculate alpha. Span generally used which denotes N(period)

alpha = 2/(span + 1)

```python
df["EWMA_12"] = df['Thousands of Passengers'].ewm(span= 12).mean()
df["EWMA_6"] = df['Thousands of Passengers'].ewm(span= 6).mean()
```

```python
df.plot()
```

Lesser the span value, greater the alpha , more weight to recent values, lesser lag, less smoothing
