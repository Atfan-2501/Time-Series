```python
df = pd.read_csv("C:/Users/2068671/Downloads/UDEMY_TSA_FINAL/UDEMY_TSA_FINAL/Data/RestaurantVisitors.csv", index_col = "date", parse_dates= True)
```

```python
df1.info()
```

We observe that the last 5 attributes are float-type inspite of the values being integers. This is because Nan cannot be accepted as integer value. Converting into integer datatype

```python
cols = ["rest1", "rest2", "rest3", "rest4", "total"]
for col in cols:
    df1[col] = df1[col].astype("int")
df1.head()
```

Step 2: Plotting the data

```python
title = "Restaurant Visitors"
ylabel = "Visitors per day"
ax = df1["total"].plot(figsize = (16, 5))
ax.set(title = title, ylabel = ylabel)
```

Highlighting the holidays on the plot

```python
ax = df1["total"].plot(figsize = (16, 5))
ax.set(title = title, ylabel = ylabel)
for x in df1[df1["holiday"] == 1].index:
    ax.axvline(x = x, color = "red", alpha = 0.3)
```

Step 3: ETS Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df1["total"])
result.plot()
plt.rcParams["figure.figsize"] = 16, 5
```

Step 4: adfuller test

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adf_result = adfuller(df1["total"], autolag= "t-stat")
adf_result
```

p-value << 0.05; Null hypothesis rejected, Data is stationary. Although there is seasonality present, adf test shows that the data is stationary. But we have seen from decomposition plot; we will consider the seasonality

Step 5: Determining the order using auto_arima

```python
from pmdarima import auto_arima
```

```python
result = auto_arima(df1["total"], seasonal= True, m = 7)
result.summary()
```

This provides an ARIMA order of (1, 0, 0) and seasonal order of (1, 0, 1, 7)

Step 6: Splitting the data into train and test

```python
df1.shape
```

```python
train = df1.iloc[ :436]
test = df1.iloc[436: ]
```

Step 7: Building the model

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = SARIMAX(train["total"], order = (1, 0, 0), seasonal_order= (1, 0, 1, 7), enforce_invertibility= False)
```

```python
results = model.fit()
results.summary()
```

Step 8: Making predictions for the test data

```python
start = len(train)
end = len(train) + len(test) - 1
predictions = results.predict(start = start, end = end, dynamic= False).rename("SARIMAX (1, 0, 0) (1, 0, 1, 7)")
```

Dynamic = False means forecast at a particular point are made using the full history upto that point

Plotting the predictions along with the test value

```python
title = "Restaurant Visitors"
ylabel = "Visitors per day"
ax = test["total"].plot(figsize = (16, 5), legend = True)
ax.set(title = title, ylabel = ylabel)
predictions.plot(legend = True)
```

Let's check how Seasonal ARIMA performs on holidays

```python
ax = test["total"].plot(figsize = (16, 5), legend = True)
ax.set(title = title, ylabel = ylabel)
predictions.plot(legend = True)
for x in test[test["holiday"] == 1].index:
    ax.axvline(x = x, color = "red", alpha = 0.3)
```

We can observe there are 4 instances where the charestic rise in visitors on a holiday is not captured using a Seasonal ARIMA, Hence we make use of exogenous variable

STEP 9: Evaluate the model 

```python
from statsmodels.tools.eval_measures import rmse, mse
```

```python
metrics = [rmse, mse]
for metric in metrics:
    error = metric(test["total"], predictions)
    print("SARIMA (1, 0, 0)(1, 0, 1, 7) {}: {}".format(str(metric).split(" ")[1], error))
```

EXOGENOUS VARIABLE (holiday)

```python
model = SARIMAX(train["total"],exog= train["holiday"],  order = (1, 0, 0), seasonal_order= (1, 0, 1, 7), enforce_invertibility= False)
results = model.fit()
results.summary()
```

```python
# Making predictions for test values
start = len(train)
end = len(train) + len(test) - 1
predictions = results.predict(start = start, end = end, exog = test["holiday"], dynamic= False).rename("SARIMAX (1, 0, 0) (1, 0, 1, 7)")
```

```python
# Plotting the predictions along with the test values
title = "Restaurant Visitors"
ylabel = "Visitors per day"
ax = test["total"].plot(figsize = (16, 5), legend = True)
ax.set(title = title, ylabel = ylabel)
predictions.plot(legend = True)
```

Lets check what modifications occur on adding the exogenous variable

```python
ax = test["total"].plot(figsize = (16, 5), legend = True)
ax.set(title = title, ylabel = ylabel)
predictions.plot(legend = True)
for x in test[test["holiday"] == 1].index:
    ax.axvline(x = x, color = "red", alpha = 0.3)
```

Evaluating the model having exogenous variable

```python
metrics = [rmse, mse]
for metric in metrics:
    error = metric(test["total"], predictions)
    print("SARIMA (1, 0, 0)(1, 0, 1, 7) {}: {}".format(str(metric).split(" ")[1], error))
```

We can observe that there is a considerable improvement in the accuracy
